```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure,xticks
%matplotlib inline

DATA_DIR = '.'
```

```python
data = pd.concat([
    pd.read_csv('./trip_advisor/xlm_r/gains.csv'),
    pd.read_csv('./toxic/xlm_r/gains.csv'),
    pd.read_csv('./prosa/xlm_r/gains.csv')
])
data.head(20)
```

```python
data.data.values
```

```python
data = data.sort_values(by=['data'])
data.head(20)
```

```python
data_mean = data.groupby(['data']).mean()
data_mean = data_mean.reset_index()
data_mean.head(10)
```

```python
data = data_mean
data.head(10)
```

```python
data.data.values
```

```python
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')


plt.plot(data.data.values,
        data.gain.values)

xticks([1000,  2500,  5000,  7500, 10000], [1000,  2500,  5000,  7500, 'MAX'])
# plt.legend(loc="upper right")
plt.xlabel('Banyak data bahasa Indonesia')
plt.ylabel('Rata-rata perbedaan F1-score baseline dengan hasil tertinggi')
plt.title('')
plt.savefig('plot.png')
plt.show()
```
