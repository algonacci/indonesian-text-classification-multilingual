```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline

DATA_DIR = '.'
```

# Load Data

```python
data = pd.DataFrame()
```

```python
total_data = [500, 1000, 2500, 5000, 7500, 11852]

for x in total_data:
    data = pd.concat([
     data,   
     pd.read_csv('{}/final_toxic_toxic_XLM_R_result_combined_{}.csv'.format(DATA_DIR, x))
    ])

print(data.shape)
data.head()
```

# Plot F1 By Total Data for each Type

```python
data.foreign_mult.unique()
```

```python
data[data['tipe'] == 'A']
```

```python
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

plt.plot(data[data['tipe'] == 'A'].total_data.values,
        data[data['tipe'] == 'A'].max_f1.values,
        
        label='Hanya B.Indo')

plt.plot(data[data['tipe'] == 'B'].total_data.values,
        data[data['tipe'] == 'B'].max_f1.values,
        
        label='Hanya B.Inggris')

plt.plot(data[(data['tipe'] == 'C') & (data['foreign_mult'] == 0.25)].total_data.values,
        data[(data['tipe'] == 'C') & (data['foreign_mult'] == 0.25)].max_f1.values,
        
        label='Campur-1:0.25')

plt.plot(data[(data['tipe'] == 'C') & (data['foreign_mult'] == 0.5)].total_data.values,
        data[(data['tipe'] == 'C') & (data['foreign_mult'] == 0.5)].max_f1.values,
        
        label='Campur-1:0.5')

plt.plot(data[(data['tipe'] == 'C') & (data['foreign_mult'] == 0.75)].total_data.values,
        data[(data['tipe'] == 'C') & (data['foreign_mult'] == 0.75)].max_f1.values,
        
        label='Campur-1:0.75')

plt.plot(data[(data['tipe'] == 'C') & (data['foreign_mult'] == 1)].total_data.values,
        data[(data['tipe'] == 'C') & (data['foreign_mult'] == 1)].max_f1.values,
        
        label='Campur-1:1')

plt.plot(data[(data['tipe'] == 'C') & (data['foreign_mult'] == 1.5)].total_data.values,
        data[(data['tipe'] == 'C') & (data['foreign_mult'] == 1.5)].max_f1.values,
        
        label='Campur-1:1.5')

plt.plot(data[(data['tipe'] == 'C') & (data['foreign_mult'] == 2)].total_data.values,
        data[(data['tipe'] == 'C') & (data['foreign_mult'] == 2)].max_f1.values,
        
        label='Campur-1:2')

plt.plot(data[(data['tipe'] == 'C') & (data['foreign_mult'] == 3)].total_data.values,
        data[(data['tipe'] == 'C') & (data['foreign_mult'] == 3)].max_f1.values,
        
        label='Campur-1:3')

plt.plot(data[(data['tipe'] == 'C') & (data['foreign_mult'] == 4)].total_data.values,
        data[(data['tipe'] == 'C') & (data['foreign_mult'] == 4)].max_f1.values,
        
        label='Campur-1:4')

plt.plot(data[(data['tipe'] == 'C') & (data['foreign_mult'] == 5)].total_data.values,
        data[(data['tipe'] == 'C') & (data['foreign_mult'] == 5)].max_f1.values,
        
        label='Campur-1:5')

plt.plot(data[(data['tipe'] == 'C') & (data['foreign_mult'] == 6)].total_data.values,
        data[(data['tipe'] == 'C') & (data['foreign_mult'] == 6)].max_f1.values,
        
        label='Campur-1:6')

plt.legend(loc="lower right")
plt.xlabel('Total Data')
plt.ylabel('F1 Score')
plt.title('Plot Prosa (XLM-R & Eng)')

plt.show()
```

# Plot all 9 experiment (by total data) side by side

```python
figure(num=None, figsize=(15, 15), dpi=80, facecolor='w', edgecolor='k')
count_plot = 0
for x in total_data:
    count_plot += 1
    plt.subplot(3, 3, count_plot)
    final_result = data[data['total_data'] == x]
    
    plt.plot(np.zeros_like(final_result[final_result['tipe'] == 'A'].max_f1.values), 
             final_result[final_result['tipe'] == 'A'].max_f1.values, 'ro', label='A')

    plt.plot(np.zeros_like(final_result[final_result['tipe'] == 'B'].max_f1.values), 
             final_result[final_result['tipe'] == 'B'].max_f1.values, 'bo', label='B')

    plt.plot(final_result[final_result['tipe'] == 'C'].foreign_mult.values, 
             final_result[final_result['tipe'] == 'C'].max_f1.values, 'go', label='C')

#     plt.legend(loc="lower right")
    plt.xlabel('Foreign Lang Multiplier')
    plt.ylabel('F1 Score')
    plt.title('{} data'.format(x))

#     plt.figtext(.7, .4, "diff = {:.5f}".format(
#         final_result[final_result['tipe'] == 'C'].f1.values[0]
#         -
#         final_result[final_result['tipe'] == 'A'].f1.values[0]))

plt.show()
```
